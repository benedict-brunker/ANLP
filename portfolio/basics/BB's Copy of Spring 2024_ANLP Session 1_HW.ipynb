Lexical attributes: scan for numbers

```python
nlp = spacy.blank("en")

doc = nlp(
    "In 1990, more than 60% of people in East Asia were in extreme poverty. "
    "Now less than 4% are."
)

for token in doc:
  if token.like_num:
    next_token = doc[token.i+1]
    if next_token.text == "%":
      print("Percentage found:", token.text)
```

### 3. Tokenizer
Create a custom tokenizer in Python that handles:
*   Contractions (e.g., "don't" → "do n't")
*   Keeps punctuation as separate tokens
*   Splits hyphenated words (e.g., "state-of-the-art" → "state of the art")

Compare its results with NLTK's word_tokenize on any sample paragraph and the following examples:
"New York-based company", "It's a beautiful day!", "https://www.example.com"

What differences do you see? What are the advantages, and limitations of each approach?

```python
def ispunct(char):
  return char in string.punctuation

def tokenizer(string):
# we initialize a cursor, which starts at the first character of the string.
# Whenever we add a token to our token_list, we'll update the cursor so that it sits at the start of the next token.
  cursor = 0
  token_list = []
# now we initialize a counter to iterate through the whole string
  for i in range(len(string)):
    start_of_word = cursor
    # if the current character is not a letter or number
    if string[i].isalnum() == False:
      # First check if it's whitespace or hyphen
      if string[i] == ' ' or string[i] == '-':
        print("Space/hyphen condition activated")
        token = string[start_of_word:i]
        token_list.append(token)
        print("new token added")
        cursor = i + 1
      # it's vs. its is a unique case
      elif string[i] =="'":

      # Then check if it's an apostrophe used for contraction
      elif string[i] == "'" and string[i+1] != 's': # this lookahead condition hopefully accounts for ' used to denote the possessive case.
        print("contraction condition activated")
        token1 = string[start_of_word:i-1]
        token2 = string[i-1:i+2]
        token_list.append(token1)
        token_list.append(token2)
        print("two new tokens added")
        cursor = i + 3 # skip the whole contracted word and succeeding whitespace
      else:
        print("punctuation mark found")
        token = string[i]
        token_list.append(token)
        cursor = i + 2 # skip the whitespace that follows the punctuation mark

      # if it's neither of those, keep the char as a separate token

    # we also need a condition for the last char in the string
    if i == len(string) - 1:
      token = string[start_of_word:i+1]
      token_list.append(token)
      return token_list
  # if function hasn't returned early then there's been an error
  print("error 1")
  return 1
```

```python
tokens = tokenizer("New York-based company")
for token in tokens:
  print(token)
```

```python
tokens2 = tokenizer("It's a beautiful day!")
for token in tokens2:
  print(token)
```
