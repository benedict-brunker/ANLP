```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the website
the_abolition_of_the_state = "https://www.marxists.org/archive/kautsky/1881/state/2-abolition.htm"
response = requests.get(the_abolition_of_the_state)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')
# The above line creates a BeautifulSoup object from the HTML content of the response.
##  It uses the 'html.parser' to parse the HTML

# Find the required information using HTML tags and attributes
title = soup.find_all('h1')[0].text # finds all<h1> tags in the HTML and takes the text of the first one (index=0) -- assumed to be the title of the page
# finds all h2 tags and returns the first one as the author of the text
author = soup.find_all('h2')[0].text
# finds all h3 tags and returns the first one as the date of publication
date_of_first_publication = soup.find_all('h3')[0].text
# finds all <em> tags and returns the first one as the original title of publication
original_publication_title = soup.find_all('em')[0].text
# finds all <strong> tags and returns the first one as the initial publisher
original_publisher  = soup.find_all('strong')[0].text


# print the scraped information
print("Title:" , title)
print("Author:", author)
print("Date of first publication:", date_of_first_publication)
print("Original title of publication:", original_publication_title)
print("Original publisher:", original_publisher)
```

```python
# extracting MIA
MIA = soup.find_all('strong')[1].text
print("Site publisher:", MIA)
```

```python
# extracting blurb
blurb = soup.find_all('strong')[2].text
print("Blurb:", blurb)
```

# Word Embeddings with Word2Vec

```python
import gensim
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize, word_tokenize
import numpy as np
import nltk
nltk.download('punkt')
import warnings
warnings.filterwarnings(action='ignore')
```

```python
# Function to process the text input and create a list of tokenized sentences
def createDataArray(text):
  # replaces newline with space to ensure the text is treated as a continuous string
  continuousText = text.replace("\n", "")
  data = [] # initializes an empty list to store the tokenized sentences
  # iterate through each sentence in the file
  for i in sent_tokenize(continuousText):
    temp = []

    # tokenize the sentence into words
    for j in word_tokenize(i):
      temp.append(j.lower())
    data.append(temp)
  return data
```

```python
data = createDataArray(text)
data
```

```python
help(gensim.models.Word2Vec)
```

```python
from nltk.corpus import stopwords
import string

stop_words = set(stopwords.words("english"))
puncts = list(string.punctuation)

def filter(data):
  filtered_sentences = []
  for sentence in data:
    filtered_words = []
    for word in sentence:
      if word not in stop_words and word not in puncts:
        filtered_words.append(word)
    filtered_sentences.append(filtered_words)
  return filtered_sentences
```

```python
filtered_data = filter(data)
filtered_data
```

```python
# Function to create a CBOW model
def createCBOWModel(data):
  cbow_model = gensim.models.Word2Vec(
      sentences=data,
      min_count=1,
      vector_size=100,
      window=5,
      sg=0
  )
  return cbow_model
```

```python
# Function to create a Skip Gram model
def CreateSkipGramModel(data):
  skipgram_model = gensim.models.Word2Vec(
      sentences=data,
      min_count=1,
      vector_size=100,
      window=5,
      sg=1
  )
  return skipgram_model
```

```python
cbow_model = createCBOWModel(filtered_data)
skipgram_model = CreateSkipGramModel(filtered_data)

# Test the CBOW model
print("CBOW Model - Words similar to 'revolution':")
similar_words_skipgram = skipgram_model.wv.most_similar("revolution", topn=5)
for word, similarity in similar_words_skipgram:
  print(f"{word}: {similarity:.4f}")
```
